#### Question 1:
Write a program to list all the inflammation csv files in the 'inflammation-data' directory and count the number of such files. An inflammation csv file has a name containing 'inflammation' and with extension '.csv'.

````python
# import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from os import listdir

#Define a function to list all the inflammation csv files and their count
def fileInfo(path):
    #create a variable to store all file names in the given path
    filenames = listdir("../DataSets/inflammation-data")
    #create a variable to store the output of a loop that returns 
    #filenames ending with .csv containing the name "inflammation"
    x =  [filename for filename in filenames if filename.endswith(".csv")
      and filename[0:12] == "inflammation"]
    #return the list of filenames and the count of those files 
    return x, len(x)

fileInfo("../DataSets/inflammation-data")
````
