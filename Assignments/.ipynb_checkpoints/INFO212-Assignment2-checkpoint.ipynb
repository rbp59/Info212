#### Question 1:
Write a program to list all the inflammation csv files in the 'inflammation-data' directory and count the number of such files. An inflammation csv file has a name containing 'inflammation' and with extension '.csv'.

````python
# import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from os import listdir

#Define a function to list all the inflammation csv files and their count
def fileInfo(path):
    #create a variable to store all file names in the given path
    filenames = listdir(path)
    #create a variable to store the output of a loop that returns 
    #filenames ending with .csv containing the name "inflammation"
    files =  [filename for filename in filenames if filename.endswith(".csv")
      and "inflammation" in filename]
    #return the list of filenames and the count of those files 
    print(files) 
    print(len(files))

fileInfo("../DataSets/inflammation-data")
````

#### Qestion 3:
Given the following list of file names: 
```
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv']
```
Write a program to plot the daily averages, maximums, and minimums of inflammation for each of the file. Your plot should have three rows each of which shows side-by-side plots of average, max, and min. The following figure illustrates the expected outcome:
![](assignment2-question3.png)

````python
# function to plot mean, max, and min data for three input files
def plotFiles(file1, file2, file3):
    #load the data files into variables
    file1 = np.loadtxt("../DataSets/inflammation-data/inflammation-02.csv",
                       delimiter = ",")
    file2 = np.loadtxt("../DataSets/inflammation-data/inflammation-04.csv",
                       delimiter = ",")
    file3 = np.loadtxt("../DataSets/inflammation-data/inflammation-06.csv",
                       delimiter = ",")
    #an array to hold the file names
    names = ["Inflammation-02.csv", "Inflammation-04.csv",
             "Inflammation-06.csv"]
    
    #an array to hold the loaded data files
    all_files = [file1, file2, file3]
    
    #loop through each file  
    for i in range(0,3,1):
        #compute arrays to hold calculated values 
        a = np.mean(all_files[i], axis = 0)
        b = np.max(all_files[i], axis = 0)
        c = np.min(all_files[i], axis = 0)
        
        #set up the plot
        layout = plt.figure(figsize=(20,12))
        #display the filenames
        plt.suptitle(names[i], fontsize = 14)
        
        #add subplots for mean, max, and min plots
        axes1 = layout.add_subplot(331) 
        axes2 = layout.add_subplot(332) 
        axes3 = layout.add_subplot(333)
        
        #plot the charts and set the labels
        axes1.plot(a) 
        axes1.set_ylabel("Average", fontsize = 12)
        axes2.plot(b) 
        axes2.set_ylabel("Maximum", fontsize = 12)
        axes3.plot(c) 
        axes3.set_ylabel("Minimum", fontsize = 12)
        
        #configure the space
        plt.subplots_adjust(left=0.2, wspace=0.25, top=0.95)
plotFiles("../DataSets/inflammation-data/inflammation-02.csv", "../DataSets/inflammation-data/inflammation-04.csv", "../DataSets/inflammation-data/inflammation-06.csv")
````

Download the data file "winemag-data-130k-v2.csv" and load it into a Pandas DataFrame using the column 0 as index. This data set contains 130k reviews for different wine verieties. Answer the following questions:
* What features do the reviews use?
* What is the description for the third view?
* Select the country, province, region_1, and region_2 columns of the records with the 0, 1, 10, and 100 index positions.
* Select all the reviews for wines made in 'Italy'. How many reviews are there for wines made in 'Italy'?
* Suppose the 'points' column represents the rating of a wine. An above-averagely good wine has a value in 'points' equal or greater than 90. Between France and Italy, which country has more reviews for its above-averagely good wines?

````python
#create a variable to hold the wine review data
wineData = pd.read_csv("../DataSets/winemag-data-130k-v2.csv", index_col = 0)

#create a variable to hold the column names
features = wineData.columns
print(features)

# return the description for the third view
description = wineData.iloc[2]["description"]
print(description)

#Select the country, province, region_1, and region_2 columns
#of the records with the 0, 1, 10, and 100 index positions.

selection = wineData.loc[[0, 1, 10, 100], 
             ['country', 'province', "region_1", "region_2"]]
print(selection)

#Select all the reviews for wines made in 'Italy'. 
#find how many reviews are there for wines made in 'Italy'
Italy = wineData[wineData['country'] == "Italy"]
print("There are ",Italy.shape[0]," reviews for wines made in Italy.")

#Find which country between France and Italy that produces
#more above-averagely good wines?

#create variables that hold country specific data
France = wineData['country'] == "France" 
Italy = wineData['country'] == "Italy" 

# Create variable holding instances when points are greater or equal to 90
criteria = wineData['points'] >= 90

# Select all cases where nationality is USA and age is greater than 50
combined = wineData[(France | Italy ) & criteria]

#separate the data from France and Italy
f = combined["country"] == "France"
i = combined["country"] == "Italy"

#print out results
print(("Italy has {0} above-averagely good wines reviews.".format(sum(i))),
      ("France has {0} above-averagely good wines reveiws.".format(sum(f))))
````
